```python
import pandas as pd
```

```python
experian = pd.read_csv('ExperianData.csv')
```

```python
experian.head(3)
```

```python
def logic(val):
    
    if val>= 0.8:
        res = 'Very Likely'
    elif val >= 0.6:
        res = 'Likely'
    elif val>=0.4:
        res = 'Somewhat Likely'
    elif val>=0.2:
        res = 'Unlikely'
    elif val>=0:
        res = "Very Unlikely"
    else: 
        res='NA'
    
    return res
```

```python
## Create a frame with the columns
dataframe_cols = experian.columns
```

```python
## Create a loop that applies the function "Logic" to the entire dataset "experian" ##

# Looks at the range of columns
for i in range(len(dataframe_cols)):
    
    # If the column starts with p_0 then create the column p_000 + i, then apply the function "logic"
    if 'p_0' in str(dataframe_cols[i]):
        experian['p_'+str(i)+'_cat'] = list(map(logic, experian[dataframe_cols[i]]))
```

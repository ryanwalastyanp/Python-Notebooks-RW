```python
import pandas as pd
df = pd.read_pickle('clean_data_for_modelling3.p')
```

```python
df.head()
```

```python
affluence_df = pd.read_csv('hub_plus_experian_affluence.csv')
registration_source_df = pd.read_csv('registration_source.csv')
first_week_completed_episodes_df = pd.read_csv('first_week_completed_episodes.csv')
average_completion_df = pd.read_csv('average_completion.csv')
svod_subscription_df = pd.read_csv('svod_subscription.csv')
```

```python
from functools import reduce
df_final = reduce(lambda left,right: pd.merge(left,right,on='viewer_id'), 
                  [df, affluence_df, registration_source_df, first_week_completed_episodes_df,
                  average_completion_df, svod_subscription_df])
```

```python
df_final.head()
```

```python
df_final = df_final.rename(columns={'case': 'registration_source'})
```

```python
df_final= df_final.fillna(0)
df_final = df_final.dropna(axis=0,how='any')
```

```python
df_final.head()
```

```python
df_final.to_pickle('clean_data_for_modelling4.p')
```

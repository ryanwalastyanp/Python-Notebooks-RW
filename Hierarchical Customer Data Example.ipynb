```python
import pandas as pd
import numpy as np

data = pd.read_csv('ds_group_session_data.csv', names = ['user_id','created','date_of_birth','gender','age','barb','postcode','town'])
data.head()
```

```python
data_h_indexed = data.set_index([data.town, data.postcode, data.user_id])
data_h_indexed = data_h_indexed.drop(['town','postcode','user_id'], axis=1)
data_h_indexed.head()
```

```python
age_cuts = [16, 24, 34, 44, 54, 64, 100]
group_names = ['16-24', '25-34', '35-44', '45-54', '55-64', '65-100']
age_bins = pd.cut(data.age, age_cuts, labels = group_names)
data = data.assign(age_group = age_bins.fillna('65-100'))
data
```

```python
age_group_dummies = pd.get_dummies(data['age_group'], prefix ="age_group_")
age_group_dummies
```

```python
gender_name = {
    'M': 0,
    'F': 1
}

data = data.assign(gender_numeric = data.gender.map(gender_name))
```

```python
data.head()
```

```python
from pandas import tseries

def correlation(data):
    y = data['age']
    x = data['gender_numeric']
    result = y.corr(x)
    return result

data.groupby('town').apply(correlation)
```

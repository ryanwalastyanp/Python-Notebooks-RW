# Import Libraries

```python
import pandas as pd
```

# Load dataframes

```python
df_startend_dates=pd.read_csv('Hub+_start_end_dates.csv')
df_clusters=pd.read_csv('cluster_simple_261018.csv')
```

# Add in Column Headers for the df_startend_dates csv

```python
df_startend_dates.columns = ['viewer_id','start_date','end_date','diff']
```

```python
df_clusters.head(10)
```

```python
# Reduce dataframe, only the viewer_id and cluster label is needed
df_clusters=df_clusters[['viewer_id','clusters']]
```

# Merge the two csv files

```python
from functools import reduce

df_clusters2 = reduce(lambda left,right: pd.merge(left,right,on='viewer_id', how='inner'), 
                  [df_startend_dates, df_clusters])
```

```python
df_clusters2.head(10)
```

```python
df_clusters2.dtypes
```
